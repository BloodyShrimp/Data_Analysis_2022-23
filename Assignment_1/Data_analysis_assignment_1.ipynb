## Exercise 2
Using scatter plot:  
a) Show the relation of square footage and price.  
b) Check how price is influenced by apartment's grade. Find other variables correlated with price.  
c) Check how good apartments are distributed over the city. First, plot zipcode versus price. Then, use apartments coordinates (longitude and lattitude) in order to show where apartments of good grade are located. 

```python
#a)
#Your code:
plt.scatter(data['sqft_living'], data['price'])
plt.ticklabel_format(style='plain')
plt.xlabel('living sqft')
plt.ylabel('price')
plt.show()

#Hint plt.scatter

# It seems that footage has some influence of the price but the discrepancies are too big for it to be the only deciding factor
```

```python
#b)
#Your code:
plt.scatter(data['grade'], data['price'])
plt.ticklabel_format(style='plain')
plt.xlabel('grade')
plt.ylabel('price')
plt.show()
plt.scatter(data['bathrooms'], data['price'])
plt.ticklabel_format(style='plain')
plt.xlabel('bathrooms')
plt.ylabel('price')
plt.show()

# It seems that price goes up with grade as well as with the number of bathrooms 
```

```python
#c)
#Your code:
plt.scatter(data['zipcode'], data['price'])
plt.ticklabel_format(style='plain')
plt.xlabel('zipcode')
plt.ylabel('price')
plt.show()

# It appears that some zipcodes have higher ranges of prices of apartments
```

```python
#Your code:
plt.figure(figsize=(15,10))
plt.scatter(data['long'], data['lat'], c=data['grade'], cmap='winter')
plt.ticklabel_format(style='plain')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar()
plt.show()

# The most expensive apartments appear to be located in the city center
```

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:
data_ideal = data.loc[(data['bedrooms'] == 3) & (data['bathrooms'] == 2) & (data['floors'] == 2)]
data_ideal.describe()

#Hint: pandas.DataFrame.loc
```

```python
#b)
#Your code:
plt.scatter(data['sqft_living'], data['price'])
plt.scatter(data_ideal['sqft_living'], data_ideal['price'])
plt.xlabel("living sqft")
plt.ylabel("price")

# Searched apartments appear to be ones between 1000 and 4000 sqr footage
```

```python
#c)
#Your code:
data_ideal['price_to_footage'] = data_ideal['price'] / data_ideal['sqft_living']

data_ideal = data_ideal.sort_values('price_to_footage', ascending=True)
data_ideal.head(5)

#Hint: .loc, .sort_values(), .index

# The apartment with best price to footage ratio with given conditions appears to be apartment number 9704 with id 7452500565
```
