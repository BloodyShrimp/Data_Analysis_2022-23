```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
```

```python
pd.set_option('display.max_columns', None)
data = pd.read_csv('Challenger_Ranked_Games.csv')
data.head()
```

```python
data.describe()
```

From the quick glance at the results from the describe function we can note that on average a League of Legends match in Challenger (the highest possible rank) lasts 1435 seconds which is about 24 minutes and the average level of a champion at the end of the game is 13.2 (players start at level 1 and maximum level is 18).

From this I draw conclusion that games at the highest level are fast-paced and often decided very quickly.

```python
blue_side_wins_graph = data['blueWins'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,textprops={'fontsize': 10},labels=['Blue Won','Red Won'],title='Win percentage',ylabel='')
```

Although some players in the community have preferences in which side of the map they deem to be better, as it influences the pick order and relative lanes positions, it would seem that there is no noticeable advantage to either side.

```python
plt.figure(figsize=(15, 6))
sns.histplot(x=data['gameDuraton'])
```

This graph is very interesting, you might notice that the duration seems to be almost a normal distribution besides two significant outliers.

Reason for the first peak is because if at least one player doesn't connect to the game before minute 3:00 the team with afk (away from keyboard) player can vote to end the game early without any punishment or loss of rank.

Second noticable peak is at around minute 15, at minute 15 the option to surrender the game becomes available and many players will choose to do so if they deem the game unwinnable.

```python
wards_deaths = sns.scatterplot(x=data['blueWardPlaced'],y=data['blueTotalMinionKills'])
```

This graph is very interesting to me and quite mysterious, as a player it would not come to me to correlate wards placed (wards give vision around the map) to minions killed (minions are creatures walking the lanes of the map fighting for each team and are main source of gold income for players). I would assume that either it is because if more wards were placed then game was longer and so more minions were killed. But the graph is quite tight compared to some other scatter plots so maybe players which care about vision more are better at securing minion kills.

```python
sns.scatterplot(y=data['blueTotalGold'],x=data['blueTotalMinionKills'])
plt.show()
```

This graph as well is quite interesting because of two distinct "wings" of points. As a player, a reason for this that I inferred might be because of we call "stomp" games. A stomp game takes place when one team quickly gains advantage with kills and at that point they are so much ahead with the amount of gold they possess that it becomes even easier to kill the enemy team. In such situations player often forgo securing objectives and killing minions and instead focus on killing enemy champions snowballing their advantage even more.

```python
sns.scatterplot(y=data['blueAvgLevel'],x=data['blueKills'])
plt.show()
```

Quite a simple graph. Kills give experience which increases champion's level so more kills equals higher level. Although the max level in a standard ranked game is 18 there seems to be matches that exceed that value. I am not sure of the reason maybe the data contained some games from other modes which allow for exceeding the max level.

```python
sns.scatterplot(y=data['blueDeath'],x=data['blueKills'])
plt.show()
```

Not a strong correlation in my opinion. Teams can have many kills and few deaths but opposite happens just as often.

```python
sns.scatterplot(y=data['blueWardkills'],x=data['blueWardPlaced'])
plt.show()
```

In games in which many wards were placed, many wards were destroyed not surprising in my opinion.

```python
sns.boxplot(y=data['blueWardPlaced'],x=data['blueDragonKills'])
plt.show()
```

Here we can see that teams that have killed a higher number of Dragons in games valued wards much more. Wards are a crucial point of securing the Dragon objective as they provide vision of the enemy team which helps with coordinated fighting.

```python
sns.boxplot(y=data['blueWardPlaced'],x=data['blueBaronKills'])
plt.show()
```

Similar results for the second objective on the map that is called Baron Nashor.

```python
sns.boxplot(y=data['blueKills'],x=data['blueFirstTower'])
plt.show()
```

Teams that took the first tower had, on average, higher kill count. Securing the first tower gives more gold than other towers and opens up the map for more free movement. This is a very good advantage to have and contributes to increasing the chances of winning.

```python
sns.boxplot(y=data['blueKills'],x=data['blueFirstBaron'])
plt.show()
```

At this graph we can look from the opposite perspective. As Baron Nashor is an objective that becomes available 20 minutes into the game, which is quite late, teams with higher kill counts had an easier time securing the objective.

```python
mosaic(data=data, index=['blueFirstTower', 'blueFirstInhibitor'])
plt.ylabel('First Inhibitor')
plt.xlabel('First Tower')
plt.show()
```

The first tower is the outer most defense structure of the enemy base and the inhibitor is one of the most inner ones. Teams that have secured the first tower more often secured the first inhibitor as well.

```python
mosaic(data=data, index=['blueFirstBlood', 'blueFirstTower'])
plt.ylabel('First Tower')
plt.xlabel('First Blood')
plt.show()
```

A similar graph teams that secured First Blood (first kill of the game) had an easier time securing the first tower.

```python
data_corr = data.drop(data.columns[[0, 1]], axis=1)
data_corr = data_corr.drop(data_corr.columns[24:], axis=1)
plt.figure(figsize=(16, 16))
graph = sns.heatmap(data_corr.corr(),annot = True,cmap="PiYG",fmt='.2f')
plt.show()
```

Here I prepared a correlation heatmap for this data set. We can note that deaths more often are correlated negatively with most of positive stats or objectives.
